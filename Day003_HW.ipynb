3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
dict = {}
locations = d['cwbopendata']['dataset']['locations']
dict[locations['locationsName']] = []

for location in locations['location']:
    dict[locations['locationsName']].append(location['locationName'])

print(dict)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
dict = {}
locations = d['cwbopendata']['dataset']['locations']
dict[locations['locationsName']] = {}

for location in locations['location']:
    dict[locations['locationsName']][location['locationName']] = {}
    weatherElement = location['weatherElement'][0]
    time = weatherElement['time'][0]
    dict[locations['locationsName']][location['locationName']][time['dataTime']] = time['elementValue']['value'] + '(' + time['elementValue']['measures']  + ')'

print(dict)
```
