### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
```

```python
import re
for d in soup.find(class_="part_list_2").find_all('h3'): 
    print(d.find('em', class_=re.compile("tag\sc_\w+") ).text, d.find(class_="date").text,d.find_all('a')[-1].text)
  
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
news = {} 
for (idx, d) in enumerate(soup.find(class_="part_list_2").find_all('h3')): 
    tag = d.find('em', class_=re.compile("tag\sc_\w+") ).text
    if not tag in news:
        news[tag]=[] 
    news[tag].append(d.find(class_="date").text + '|' + d.find_all('a')[-1].text) 

print("依照字典格式將類別一起儲存"")
print(news) 
   
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
max_categoty = ""
max_count = 0
for k, v in news.items():
    if len(v) > max_count:
        max_count = len(v)
        max_categoty = k

print(max_categoty)        
```
