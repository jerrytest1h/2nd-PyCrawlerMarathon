## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 每個兩秒鐘自動往下滑
import time
for i in range(10):
    time.sleep(3)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
news = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    news.append(  d.find(class_="date").text + "|" + d.find_all('a')[-1].text  )
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)   

print(news)    
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
from datetime import datetime
# d1 = datetime.now()
# print(d1.time)
# def f(dt):
#      print(dt.time)
#      return datetime.time(21, 20) <= dt.time() <= datetime.time(23, 20)

for n in news:
    _n = n.split('|')
    dt = datetime.strptime(_n[0], "%Y/%m/%d %H:%M")
    if dt.hour >= 5 and dt.hour <= 7:
        print(n)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''

# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

```

```python
from selenium.webdriver.support.ui import Select
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2020')

selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('3')

selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value('14')

browser.find_element_by_id('button').click()

# 每個兩秒鐘自動往下滑
import time
time.sleep(3)  
```

```python
for i in range(100):
    time.sleep(3) 
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
news2 = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    news2.append(  d.find(class_="date").text + "|" + d.find_all('a')[-1].text  )
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)   

print(news2)    
```

```python
from datetime import datetime
for n in news2:
    _n = n.split('|')
    dt = datetime.strptime(_n[0], "%Y/%m/%d %H:%M") 
    if dt.strftime("%d") == "14":
        if dt.hour >= 15 and dt.hour <= 16:
            print(n)
```
